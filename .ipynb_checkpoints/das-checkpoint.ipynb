```python
df["SEX"]  = df["SEX"].replace({2:0})
```

```python
df["HIPERTENSION"]  = df["HIPERTENSION"].replace({2:0, 98:0})
```

```python
df["RENAL_CHRONIC"]  = df["RENAL_CHRONIC"].replace({2:0,98:0})
```

```python
df["TOBACCO"]  = df["TOBACCO"].replace({2:0,98:0})
```

```python
df["ICU"]  = df["ICU"].replace({2:0,97:0, 99:0})
```

```python
df["survive"] = df["DATE_DIED"].apply(
    lambda x: 1 if x== '9999-99-99' else 0)
```

### NOW start visualtion 

# Gender Distribution fro total dataset

```python
sns.countplot(x=df['SEX'], legend=True, palette = ["Blue", "Pink"])
plt.xticks([0,1],["Male", "Female"])
plt.xlabel('Gender', fontweight = "bold")
plt.ylabel('Count',fontweight = "bold")
plt.title('Gender Distribution', fontweight = "bold")
plt.show()
```

# Gender VS Death

```python
total_death_male = df[(df["SEX"] == 0) & (df["survive"] == 0)].shape[0]
print("total_death_male",total_death_male)
total_death_Female = df[(df["SEX"] != 0) & (df["survive"] == 0)].shape[0]
print("total_death_Female",total_death_Female)
```

```python
sns.countplot(data  = df[df["survive"] == 0], x = 'SEX', palette = ["blue", "pink"])
plt.xticks([0,1],["Male","Female"])
plt.xlabel("Gender", fontweight = "bold")
plt.ylabel("Count", fontweight = "bold")
plt.title("Gender vs Death", fontweight = "bold")
plt.show()
```

# Age distribution

```python
sns.histplot(x = "AGE", data = df, bins= 20, kde = True, color = "purple")
plt.title("Age Distribution", fontweight = "bold")
plt.xlabel("Age", fontweight = "bold")
plt.ylabel("no of people", fontweight = "bold")
plt.show()
```
